# Console-Based Calculator

````python
# Define functions for each arithmetic operation

def add(x, y):
    """Adds two numbers."""
    return x + y

def subtract(x, y):
    """Subtracts two numbers."""
    return x - y

def multiply(x, y):
    """Multiplies two numbers."""
    return x * y

def divide(x, y):
    """Divides two numbers, handles division by zero."""
    if y == 0:
        return "Error! Division by zero!"
    return x / y

def console_calculator():
    """Runs the main console calculator loop."""
    print("\nWelcome to the Simple Console Calculator!")

    while True:
        # Display the operation menu to the user
        print("\n--- Select Operation ---")
        print("1. Add")
        print("2. Subtract")
        print("3. Multiply")
        print("4. Divide")
        print("5. Exit")
        print("------------------------")

        choice = input("Enter choice (1/2/3/4/5): ")

        # Check if the user wants to exit
        if choice == '5':
            print("Thank you for using the calculator. Goodbye!")
            break

        # Validate the operation choice
        if choice in ('1', '2', '3', '4'):
            try:
                # Get valid numerical input from the user
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
            except ValueError:
                # Handle invalid input if numbers are not entered
                print("Invalid input. Please enter numbers only.")
                continue # Continue to the next iteration of the loop

            # Perform the chosen operation and display the result
            if choice == '1':
                print(f"{num1} + {num2} = {add(num1, num2)}")
            elif choice == '2':
                print(f"{num1} - {num2} = {subtract(num1, num2)}")
            elif choice == '3':
                print(f"{num1} * {num2} = {multiply(num1, num2)}")
            elif choice == '4':
                result = divide(num1, num2)
                print(f"{num1} / {num2} = {result}")
        else:
            # Handle invalid operation choice
            print("Invalid Input. Please enter a choice between 1 and 5.")

# To run the console calculator, uncomment the line below:
# console_calculator()
````

# GUI Calculator using Tkinter

````python
import tkinter as tk
from tkinter import messagebox

class GUICalculator:
    """A simple GUI calculator using Tkinter."""

    def __init__(self, master):
        self.master = master
        master.title("Simple Calculator")
        master.geometry("300x250") # Set a fixed window size for better layout consistency
        master.resizable(False, False) # Prevent resizing

        # Variable to store the result displayed on the GUI
        self.result_var = tk.StringVar()
        self.result_var.set("0") # Initialize with 0

        # Create GUI components
        self.create_widgets()

    def create_widgets(self):
        """Creates all the widgets for the calculator GUI."""

        # Entry widgets for number input
        self.num1_label = tk.Label(self.master, text="Number 1:")
        self.num1_label.grid(row=0, column=0, padx=5, pady=5, sticky='w')
        self.num1_entry = tk.Entry(self.master, width=20)
        self.num1_entry.grid(row=0, column=1, columnspan=3, padx=5, pady=5)

        self.num2_label = tk.Label(self.master, text="Number 2:")
        self.num2_label.grid(row=1, column=0, padx=5, pady=5, sticky='w')
        self.num2_entry = tk.Entry(self.master, width=20)
        self.num2_entry.grid(row=1, column=1, columnspan=3, padx=5, pady=5)

        # Buttons for operations
        tk.Button(self.master, text="Add", command=lambda: self.perform_operation('add'), width=7).grid(row=2, column=0, padx=5, pady=5)
        tk.Button(self.master, text="Subtract", command=lambda: self.perform_operation('subtract'), width=7).grid(row=2, column=1, padx=5, pady=5)
        tk.Button(self.master, text="Multiply", command=lambda: self.perform_operation('multiply'), width=7).grid(row=3, column=0, padx=5, pady=5)
        tk.Button(self.master, text="Divide", command=lambda: self.perform_operation('divide'), width=7).grid(row=3, column=1, padx=5, pady=5)

        # Result display
        self.result_label_text = tk.Label(self.master, text="Result:")
        self.result_label_text.grid(row=4, column=0, padx=5, pady=5, sticky='w')
        self.result_display = tk.Label(self.master, textvariable=self.result_var, font=('Arial', 14, 'bold'), fg='blue', anchor='w')
        self.result_display.grid(row=4, column=1, columnspan=3, padx=5, pady=5, sticky='we')

        # Clear button
        tk.Button(self.master, text="Clear", command=self.clear_fields, width=7).grid(row=5, column=0, padx=5, pady=5)

    def get_numbers(self):
        """Attempts to get and convert numbers from entry fields. Handles errors."""
        try:
            num1 = float(self.num1_entry.get())
            num2 = float(self.num2_entry.get())
            return num1, num2
        except ValueError:
            messagebox.showerror("Input Error", "Please enter valid numbers in both fields.")
            return None, None

    def perform_operation(self, operation):
        """Performs the selected arithmetic operation."""
        num1, num2 = self.get_numbers()
        if num1 is None or num2 is None: # If get_numbers returned an error
            self.result_var.set("Error")
            return

        result = None
        if operation == 'add':
            result = num1 + num2
        elif operation == 'subtract':
            result = num1 - num2
        elif operation == 'multiply':
            result = num1 * num2
        elif operation == 'divide':
            if num2 == 0:
                messagebox.showerror("Division Error", "Cannot divide by zero!")
                self.result_var.set("Error")
                return
            result = num1 / num2

        self.result_var.set(str(result))

    def clear_fields(self):
        """Clears the input fields and result display."""
        self.num1_entry.delete(0, tk.END)
        self.num2_entry.delete(0, tk.END)
        self.result_var.set("0")

# Main part of the script to run the GUI calculator
# To run the GUI calculator, uncomment the lines below:
# if __name__ == "__main__":
#     root = tk.Tk()
#     my_calculator = GUICalculator(root)
#     root.mainloop()
````
